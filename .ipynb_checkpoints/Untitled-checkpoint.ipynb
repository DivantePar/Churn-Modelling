```python
plt.figure(figsize=(14,14))

for i, col in enumerate(list(the_df.columns)):
    if the_df[col].dtype == "int64" or the_df[col].dtype == "object":
        plt.subplot(4,3,i+1)
        sns.countplot(hue=the_df['Exited'], x=the_df[col], data=the_df)

plt.subplots_adjust(wspace=0.5, hspace=0.75)
plt.show()
```
